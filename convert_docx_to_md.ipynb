```python
import mammoth
import os

# Path ke file DOCX
docx_path = "TUGAS AKHIR HarisW (1).docx"
md_path = docx_path.replace(".docx", ".md")

# Skip gambar saat convert
def ignore_image(image):
    return {}

with open(docx_path, "rb") as docx_file:
    result = mammoth.convert_to_markdown(
        docx_file,
        convert_image=mammoth.images.img_element(ignore_image)
    )
    md_content = result.value
    messages = result.messages

# Hapus sisa tag img kosong
import re
md_content = re.sub(r'!\[[^\]]*\]\([^)]*\)', '', md_content)

# Simpan ke file .md
with open(md_path, "w", encoding="utf-8") as md_file:
    md_file.write(md_content)

print(f"Converted: {docx_path} -> {md_path}")
print(f"Size: {os.path.getsize(md_path):,} bytes")
if messages:
    print(f"\nWarnings ({len(messages)}):")
    for m in messages[:10]:
        print(f"  - {m}")
```
